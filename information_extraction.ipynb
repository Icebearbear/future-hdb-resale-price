```python
def get_floor(row):
    floors = row['storey_range'].split(" TO ")
    start = int(floors[0])
    end = int(floors[1])

    return pd.Series([start, end])

def get_room(row):
    room = row['flat_type'].split(" ")[0]
    if room.isnumeric():
        room = int(room)
    return room
```

```python
df[['floor-range-start', 'floor-range-end']]  = df.apply(get_floor, axis=1)
df
```

```python
df['rooms'] = df.apply(get_room, axis=1)
```

```python
## 3. What is the highest and lowest resale price in each town 
##    with number of room ranges between 3 and 4
towns = df['town'].unique()
upper_target_room = 4
lower_target_room = 3
results_3 = []

for town in towns:
    temp = []
    for i, row in df.iterrows():
        if row['town'] == town and isinstance(row['rooms'], int):
            if row['rooms'] >= lower_target_room and row['rooms'] <= upper_target_room:
                temp.append(row) 
    temp = pd.DataFrame(temp)
    if temp.shape[0] > 0:
        temp.sort_values(by='resale_price')
        results_3.append(temp.iloc[0])
        results_3.append(temp.iloc[-1])

results_3 = pd.DataFrame(results_3)
print("The highest and lowest resale price \
      in each town with number of room ranges between 10 and 15 :", results_3)
```
